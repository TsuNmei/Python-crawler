```python
from bs4 import BeautifulSoup
import requests
```

```python
r = requests.get("https://www.ptt.cc/bbs/MobileComm/index.html") #將此頁面的HTML GET下來
print(r.text) #印出HTML
```

```python
soup = BeautifulSoup(r.text,"html.parser")
element = soup.select("div.title a")
```

```python
ls = []
for e in element:
    ls.append(e.text)
```

```python
ls[1] = ls[1][4:]
ls
```
